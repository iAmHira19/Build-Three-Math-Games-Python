```python
# Install necessary libraries
!pip install matplotlib numpy

# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, FloatSlider

# Function to plot projectile motion
def projectile_game(a, b, c):
    g = 9.81  # Gravity
    t = np.linspace(0, 10, 500)
    x = a * t
    y = b * t - 0.5 * g * t**2 + c

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'Path: y = {a}x - 0.5*{g}*t^2 + {c}')
    plt.axhline(0, color='black', lw=2)
    plt.title('Projectile Game')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.show()

# Interactive widgets for sliders
interact(
    projectile_game,
    a=FloatSlider(min=-10, max=10, step=0.1, value=1),
    b=FloatSlider(min=-10, max=10, step=0.1, value=5),
    c=FloatSlider(min=-10, max=10, step=0.1, value=0)
)

# Level 2 (without sliders)
def projectile_game_level2(a, b, c):
    g = 9.81  # Gravity
    t = np.linspace(0, 10, 500)
    x = a * t
    y = b * t - 0.5 * g * t**2 + c

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'Path: y = {a}x - 0.5*{g}*t^2 + {c}')
    plt.axhline(0, color='black', lw=2)
    plt.title('Projectile Game - Level 2')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.show()

# User inputs for Level 2
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))
projectile_game_level2(a, b, c)
```
