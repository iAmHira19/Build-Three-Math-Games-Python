```python
# Install necessary libraries
!pip install matplotlib numpy

# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Function to generate random points and plot them
def scatter_plot_game():
    # Generate random points
    x = np.random.randint(0, 100, 10)
    y = np.random.randint(0, 100, 10)

    # Plot the points
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='blue', label='Random Points')
    plt.title('Scatter Plot Game')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.show()

    # Collect user input
    print("Guess the coordinates of the points!")
    for i in range(len(x)):
        user_x = int(input(f"Point {i+1} - Enter x coordinate: "))
        user_y = int(input(f"Point {i+1} - Enter y coordinate: "))
        if (user_x == x[i]) and (user_y == y[i]):
            print("Correct!")
        else:
            print(f"Incorrect! The point was ({x[i]}, {y[i]})")

scatter_plot_game()
```
